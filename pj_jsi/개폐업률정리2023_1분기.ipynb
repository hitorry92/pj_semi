```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import PublicDataReader as pdr
```

```python
a = pd.read_csv("./2023/2023년_1분기_외식업_개폐업수(률).csv", encoding = "ansi", header = 1)
```

```python
b = pd.concat([a.iloc[:,[0,1]],a.iloc[:,-4:]], axis = 1)
b
```

```python
c = b.drop(index =[0])
c.columns = ['시군구명', '서비스업종','개업수', '폐업수', '개업률', '폐업률']
c = c[~(c["시군구명"].str.contains("서울시"))]
```

```python
c["시군구명"] = c["시군구명"].str.replace(pat = "하위 메뉴 펼침",repl = "")
c["서비스업종"] = c["서비스업종"].str.replace(pat = "외식업/ ",repl = "")
c
```

```python
c["서비스업종"].unique()
```

```python
d = ["한식음식점", "중식음식점", "일식음식점", "양식음식점", "제과점", "패스트푸드점", "치킨전문점", "분식전문점", "호프-간이주점", "커피-음료"]
e = ["한식", "중식", "일식", "기타 외국식", "제과제빵", "패스트푸드", "치킨", "분식", "주점", "커피"]
f = dict(zip(d,e))
f
```

```python
c["서비스업종"] = c["서비스업종"].replace(f)
if "분기" in c.columns:
    c.drop(columns = "분기", inplace= True)
c.insert(loc = 0, column = "분기", value = "1")
c
```

```python
c.to_csv("./개폐업률정리2023_1분기.csv", encoding="ansi", index = False)
```
