# `2. 3월, 6월, 9월, 12월을 분기 대표로 할거면 # 풀어`

```python
a = a[a["데이터기준년월"].str[-2:].isin(["03", "06", "09", "12"])]
for i,v in enumerate(["03", "06", "09", "12"]):
    a.loc[a["데이터기준년월"].str[-2:] == v, "분기"] = i+1
a["분기"] = a["분기"].astype(dtype = int)
a
```

```python
a["매출등급"] = a['매출등급'].astype(dtype = int)
```

```python
a["매출등급"] = a['매출등급'].replace(to_replace = {1: 5, 2: 4, 3: 3, 4: 2, 5: 1})
```

```python
a
```

```python
b = a[["분기","서비스업종코드","시군구명","매출등급"]].groupby(by = ["분기","시군구명","서비스업종코드"]).mean().round(0).reset_index()
```

```python
c = 코드변환.drop_duplicates(subset = "서비스업종코드")
d = b["서비스업종코드"].map(dict(zip(c["서비스업종코드"],c["업종"])))
b.insert(loc = 3, column = "서비스업종", value = d)
```
