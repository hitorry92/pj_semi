# `구와 행정동,법정동 코드 뽑아내기`

```python
행_법정동코드 = pdr.code_hdong_bdong()
행_법정동코드 = 행_법정동코드.loc[행_법정동코드['말소일자'] == ""]
행_법정동코드 = 행_법정동코드.astype(dtype = str)
행_법정동코드
```

### `행정동코드, 법정동코드 길이 8로 맞춰주기`

```python
행_법정동코드["법정동코드_8"] = 행_법정동코드["법정동코드"].str[:-2]
행_법정동코드["행정동코드_8"] = 행_법정동코드["행정동코드"].str[:-2]
행_법정동코드
```

### `행_법정동코드에서 필요한 컬럼만 가져와서 첫줄(서울특별시) 제거하기`

```python
a = 행_법정동코드[["시군구코드", "시군구명","행정동코드_8","법정동코드_8"]].drop(index = "0")
a
```

### `상권영역 str 바꿔주기`

```python
상권영역_f_xy = 상권영역_f_xy.astype(dtype = str)
```

### `상권영역에서 필요한 컬럼만 가져오기`

```python
상권영역_f_xy.columns
```

```python
b = 상권영역_f_xy[["상권_코드","상권_구분_코드","상권_구분_코드_명","행정동_코드",'folium 위도', 'folium 경도']]
b
```

### `행_법정동코드와 상권영역 필요한 컬럼들 끼리 결합하기`

```python
상권_구_행_법_코드 = pd.merge(left = b, right = a, left_on = "행정동_코드", right_on = "행정동코드_8", how = "left").drop(columns = "행정동_코드" )
상권_구_행_법_코드[["folium 위도","folium 경도"]] = 상권_구_행_법_코드[["folium 위도","folium 경도"]].astype(float)
```

```python
a = 상권_구_행_법_코드[["상권_구분_코드","상권_구분_코드_명","시군구명","folium 위도","folium 경도"]].groupby(by = ["시군구명","상권_구분_코드_명"])
```

```python
for i,df in a:
    print(i,df)
```

```python
for i,df in a:
  dict_1 = []
  list_2 = df[["folium 경도","folium 위도"]].values.tolist()
  dict_1.append({
    "type": "Feature",
    "id": f"{i[1]}",
    "properties": {
      "name": f"{i[1]}",
    },
    "geometry": {
      "type": "Polygon",
      "coordinates": [
        list_2
      ]
    }
  })     
  b = {
  "type": "FeatureCollection",
  "features": dict_1
  }

  with open(f"./임시/{i}.json", "w") as file:
      json.dump(b, file)
  
```

### `추정 매출에 구, 행정동, 법정동 코드 결합하기`

```python
추정매출_결합 = pd.merge(left = 추정매출, right = 상권_구_행_법_코드, how = "left")
```
