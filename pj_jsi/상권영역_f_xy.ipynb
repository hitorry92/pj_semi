```python
from pyproj import Transformer

# EPSG 5181 좌표계와 UTM 좌표계 간의 변환을 위한 Transformer 객체 생성
transformer = Transformer.from_crs("EPSG:5181", "EPSG:4326", always_xy=False)  # 여기서 4326는 folium 좌표계을 나타냄
# always_xy = False가 디폴트 값으로 입력으로(위도,경도) 출력으로(위도, 경도)
# always_xy = True는 입력으로(경도,위도) 출력으로(경도, 위도)
# EPSG 5181 좌표를 UTM 좌표계로 변환
l = []
for x,y in zip(상권영역["엑스좌표_값"],상권영역["와이좌표_값"]):
    #utm_x, utm_y = transformer.transform(x, y)
    l.append(transformer.transform(y, x))
    #print(f"folium 좌표: ({utm_x}, {utm_y})")
f_xy = pd.DataFrame(data = l, columns = ["folium 위도", "folium 경도"])
```

```python
import folium
import pandas as pd

# 상권 영역 데이터 (이 부분은 실제 데이터로 대체해야 합니다.)
# 상권영역 = pd.read_csv("상권영역_데이터.csv")

# EPSG 5181 좌표계와 UTM 좌표계 간의 변환을 위한 Transformer 객체 생성
transformer = Transformer.from_crs("EPSG:5181", "EPSG:4326")

# 좌표 변환
l = []
for x, y in zip(상권영역["엑스좌표_값"], 상권영역["와이좌표_값"]):
    l.append(transformer.transform(y, x))
f_xy= pd.DataFrame(data = l, columns = ["folium 위도", "folium 경도"])

# 지도 생성
map_center = [f_xy["folium 위도"].mean(), f_xy["folium 경도"].mean()]  # 평균 좌표로 중심 설정
m = folium.Map(location=map_center, zoom_start=12)

# 변환된 좌표를 이용하여 지도에 마커 추가
for _, row in f_xy.iterrows():
    folium.Marker([row["folium 위도"], row["folium 경도"]]).add_to(m)

# 지도를 HTML 파일로 저장
m.save("folium_map.html")
```
