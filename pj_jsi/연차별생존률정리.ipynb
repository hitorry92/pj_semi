# `연차별생존률 정리하기`

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import PublicDataReader as pdr
```

```python
연차별생존률 = pd.read_csv("./230825.x값 분기.구.업종_연차별 생존률_1차수정.csv", encoding = "ansi")
```

```python
d = ["한식음식점", "중식음식점", "일식음식점", "양식음식점", "제과점", "패스트푸드점", "치킨전문점", "분식전문점", "호프-간이주점", "커피-음료"]
e = ["한식", "중식", "일식", "기타 외국식", "제과제빵", "패스트푸드", "치킨", "분식", "주점", "커피"]
f = dict(zip(d,e))

c = []
for i in range(0,440,11):
    a = 연차별생존률.iloc[:,i:i+11]
    b = pd.concat(objs = [a.iloc[:,:2],a.iloc[:,8:11]], axis = 1)
    b.columns = ['시군구명', '서비스업종','1년생존률', '3년생존률', '5년생존률']
    b["서비스업종"] = b["서비스업종"].replace({"외식업/":f"{b.iloc[5,1]}"})
    b["서비스업종"] = b["서비스업종"].replace(f)
    b.dropna(axis = 0, how = "any", inplace = True)
    b.insert(loc = 0, column = "분기", value = f"{a.columns[2][6]}")
    c.append(b)
print(c)
```

```python
a = c[0]
b = c[1]
d = pd.concat(objs = [a,b])
for i in c[2:]:
    d = pd.concat(objs = [d,i])
```

```python
연차별생존률정리 = d.reset_index(drop = True).astype(str)
연차별생존률정리
```

```python
연차별생존률정리.to_csv("./연차별생존률정리.csv", encoding="ansi", index = False)
```
