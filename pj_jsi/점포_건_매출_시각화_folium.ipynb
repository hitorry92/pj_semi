```python
for i,v in df1.groupby(by = "서비스업종"):
    df2 = v
    loc = [37.4705, 126.5398]

    # 지도 정의
    m = folium.Map(location=loc, zoom_start=11)

    # choropleth 맵 그리기
    folium.Choropleth(
        geo_data=geo_data,
        data=df2,  # 데이터프레임 전체를 사용해야 합니다.
        columns=['시군구명', "업종별_점포당_매출_금액"],  # 사용할 열 이름
        fill_color='PuRd',
        key_on='feature.id'
    ).add_to(m)
    for j,k in e.items():
        text_icon = folium.DivIcon(
        icon_size=(100,20),  # DivIcon의 크기 조절
        icon_anchor=(0,0),  # 텍스트 위치 조절
        html=f'<div style="font-size: 9pt; color: black;">{j}</div>'  # 표시할 텍스트
        )
        folium.Marker(
            location = k,
            icon = text_icon,
        ).add_to(m)
    
    folium.LatLngPopup().add_to(m)
    # 생성한 지도 저장
    m.save(f'임시/{i}_점포당_매출_금액.html')
```
