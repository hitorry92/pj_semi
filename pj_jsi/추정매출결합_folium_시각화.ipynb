```python
geo_path = './임시/('강남구', '골목상권').json'
geo_data = json.load(open('./임시/('강남구', '골목상권').json', encoding='utf-8'))
```

```python
for i,v in 시각화시작.groupby(by = ["분기", "서비스업종"]):
    a = v["시군구명"].value_counts().to_frame()
    # 위치 파라미터 설정
    loc = [37.5502, 126.982] # 위도(N), 경도(E)
    data_size = len(v)
    map_seoul = folium.Map(location=loc, zoom_start=12)
    map_seoul.choropleth(geo_data=geo_data,
              data = a['count'],
              columns=[a.index, a['count']],
              fill_color='PuRd',
              key_on='feature.id')
    map_seoul.save(f"{i}.html")
```

```python
for i,v in 시각화시작.groupby(by = ["분기", "서비스업종"]):
    a = v["시군구명"].value_counts().to_frame()
    # 위치 파라미터 설정
    loc = [37.5502, 126.982] # 위도(N), 경도(E)
    data_size = len(v)
    map_seoul = folium.Map(location=loc, zoom_start=12)
    map_seoul.choropleth(geo_data=geo_data,
              data = a['count'],
              columns=[a.index, a['count']],
              fill_color='PuRd',
              key_on='feature.id')
    for j in ("2", "3", "4"):
        b = v[v["매출등급"] == f"{j}.0"]
        data_size = len(b)
        for k in range(data_size):
            if j == "2":
                folium.Marker(list(v.iloc[k][["folium 위도", "folium 경도"]]),
                              popup=v.iloc[k][['상권_코드_명']],
                              icon=folium.Icon(color='purple')).add_to(map_seoul)
            elif j == "3":
                folium.Marker(list(v.iloc[k][["folium 위도", "folium 경도"]]),
                              popup=v.iloc[k][['상권_코드_명']],
                              icon=folium.Icon(color='green')).add_to(map_seoul)
            elif j == "4":
                folium.Marker(list(v.iloc[k][["folium 위도", "folium 경도"]]),
                              popup=v.iloc[k][['상권_코드_명']],
                              icon=folium.Icon(color='blue')).add_to(map_seoul)
    map_seoul.save(f"{i}.html")
```
