# `평균영업기간 정리하기`

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import PublicDataReader as pdr
```

```python
평균영업기간 = pd.read_csv("./230825.x값 분기.구.업종_평균영업기간_1차수정.csv", encoding = "ansi")
```

```python
평균영업기간.head()
```

```python
d = ["한식음식점", "중식음식점", "일식음식점", "양식음식점", "제과점", "패스트푸드점", "치킨전문점", "분식전문점", "호프-간이주점", "커피-음료"]
e = ["한식", "중식", "일식", "기타 외국식", "제과제빵", "패스트푸드", "치킨", "분식", "주점", "커피"]
f = dict(zip(d,e))

c = []
for i in range(0,320,8):
    a = 평균영업기간.iloc[:,i:i+8]
    b = pd.concat(objs = [a.iloc[:,:2],a.iloc[:,6:8]], axis = 1)
    b.columns = ['시군구명', '서비스업종',"최근10년영업기간", "최근30년영업기간"]
    b["서비스업종"] = b["서비스업종"].replace({"외식업/":f"{b.iloc[5,1]}"})
    b["서비스업종"] = b["서비스업종"].replace(f)
    b.dropna(axis = 0, how = "any", inplace = True)
    b.insert(loc = 0, column = "분기", value = f"{a.columns[2][6]}")
    b = b.drop(labels = 2 ,axis = 0)
    c.append(b)
print(c)
```

```python
a = c[0]
b = c[1]
d = pd.concat(objs = [a,b])
for i in c[2:]:
    d = pd.concat(objs = [d,i])
```

```python
평균영업기간정리 = d.reset_index(drop = True).astype(str)
평균영업기간정리
```

```python
평균영업기간정리.to_csv("./평균영업기간정리.csv", encoding="ansi", index = False)
```
